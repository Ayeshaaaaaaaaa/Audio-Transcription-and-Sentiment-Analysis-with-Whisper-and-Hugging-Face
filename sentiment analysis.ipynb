```python
import whisper
import requests

# Load Whisper model for transcription
model = whisper.load_model("medium")

# Transcribe audio file
transcription_result = model.transcribe("/content/Power_English_Update.mp3", language="en",temperature=0.6,verbose=True )
transcribed_text = transcription_result['text']
print("Transcribed Text:", transcribed_text)
```

```python
import requests
import time

# Hugging Face Inference API URL for sentiment analysis
API_URL = "https://api-inference.huggingface.co/models/distilbert-base-uncased-finetuned-sst-2-english"
headers = {"Authorization": f"Bearer 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"}

# Function to perform sentiment analysis
def sentiment_analysis(text):
    payload = {"inputs": text}
    response = requests.post(API_URL, headers=headers, json=payload)
    return response.json()

# Chunk the text and truncate tokens if necessary
def chunk_text(text, max_length=512):
    tokens = text.split()  # Tokenize text by spaces (for simple chunking)

    # Ensure chunks are within the max length limit
    chunks = [tokens[i:i + max_length] for i in range(0, len(tokens), max_length)]

    # Join tokens back into sentences
    return [" ".join(chunk) for chunk in chunks]

# Analyze each chunk of text
def analyze_chunks(chunks):
    results = []
    for chunk in chunks:
        if len(chunk) > 512:  # Ensure the chunk length is within limit
            chunk = chunk[:512]
        result = sentiment_analysis(chunk)
        results.append(result)
    return results

# Define maximum length based on model's limit (512 tokens)
max_length = 512

# Assuming `transcribed_text` is already defined
text_chunks = chunk_text(transcribed_text, max_length)

# Perform sentiment analysis on each chunk
sentiment_results = analyze_chunks(text_chunks)

# Print results for each chunk
for i, result in enumerate(sentiment_results):
    print(f"Chunk {i + 1} Sentiment Analysis Result:", result)
```
